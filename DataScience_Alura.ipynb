```python
import pandas as pd
from sklearn.linear_model import LinearRegression # Importa o modelo de regressão
import os # Para verificar o arquivo

# --- Passo 1: Carregar o DataFrame original e treinar um modelo (Exemplo) ---
file_path = "/content/Preços_de_casas.csv"

try:
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Arquivo de treinamento não encontrado: {file_path}")

    df_treino = pd.read_csv(file_path)
    if 'Id' in df_treino.columns:
        df_treino = df_treino.drop(columns = "Id")
    print("Dataset de treino carregado com sucesso!")

    # --- VERIFICAÇÃO IMPORTANTE: Exibir as colunas do DataFrame ---
    print("\nColunas disponíveis no DataFrame de treino:")
    print(df_treino.columns.tolist())
    # --- FIM DA VERIFICAÇÃO ---


    # Pre-processamento para a variável 'qualidade_da_cozinha'
    # VERIFIQUE AQUI O NOME EXATO DA SUA COLUNA DE QUALIDADE DA COZINHA!
    # Por exemplo, se na lista acima aparecer 'Qualidade_Cozinha', mude 'qualidade_da_cozinha' para 'Qualidade_Cozinha'
    coluna_qualidade_cozinha = 'qualidade_da_cozinha' # <--- AJUSTE AQUI SE O NOME DA COLUNA FOR DIFERENTE NO SEU CSV

    if coluna_qualidade_cozinha in df_treino.columns:
        df_treino = pd.get_dummies(df_treino, columns=[coluna_qualidade_cozinha], drop_first=True, dtype=int)
        print(f"Variável '{coluna_qualidade_cozinha}' transformada com sucesso em variáveis dummy.")
        print("Novas colunas após one-hot encoding:", [col for col in df_treino.columns if 'qualidade_da_cozinha' in col])
    else:
        print(f"AVISO: A coluna '{coluna_qualidade_cozinha}' não foi encontrada no dataset. Ignorando a transformação.")


    # Definir variáveis explicativas (X) e variável resposta (y)
    # As variáveis aqui devem ser EXATAMENTE as que seu modelo real usou no treino.
    # Se você não usou 'existe_segundo_andar' ou 'qualidade_da_cozinha_Excelente', remova-as daqui.
    features_para_modelo = [
        'quantidade_banheiros',
        'area_primeiro_andar',
        'existe_segundo_andar',
        'qualidade_da_cozinha_Excelente' # Esta será criada pelo get_dummies
    ]

    # Filtra as features que realmente existem no DataFrame após o pré-processamento
    available_features = [f for f in features_para_modelo if f in df_treino.columns]

    # Se a 'qualidade_da_cozinha' original não estava presente e não criou a dummy,
    # precisamos garantir que 'qualidade_da_cozinha_Excelente' não esteja em available_features
    if 'qualidade_da_cozinha_Excelente' not in df_treino.columns and 'qualidade_da_cozinha_Excelente' in available_features:
        available_features.remove('qualidade_da_cozinha_Excelente')

    # Verificar se as features necessárias para a nova casa estão disponíveis
    required_new_house_features = [
        'quantidade_banheiros', 'area_primeiro_andar',
        'existe_segundo_andar', 'qualidade_da_cozinha_Excelente'
    ]
    # Certifica-se de que estamos apenas selecionando as que foram realmente usadas no treino
    features_para_prever_nova_casa = [f for f in required_new_house_features if f in available_features]


    X_treino = df_treino[available_features]
    y_treino = df_treino['preco_de_venda']

    # Treinar o modelo de regressão linear
    modelo_regressao = LinearRegression()
    modelo_regressao.fit(X_treino, y_treino)
    print("\nModelo de regressão treinado com sucesso com as features:", available_features)

except FileNotFoundError as e:
    print(f"ERRO: {e}")
    print("Certifique-se de que o arquivo de treino 'Preços_de_casas.csv' está no local correto.")
    exit()
except KeyError as e:
    print(f"ERRO: A coluna '{e}' não foi encontrada no dataset 'df_treino'. Verifique os nomes das colunas e o pré-processamento.")
    exit()
except Exception as e:
    print(f"Ocorreu um erro inesperado ao carregar o arquivo ou treinar o modelo: {e}")
    exit()


# --- Passo 2: Preparar os dados da nova casa para previsão ---
# Crie um DataFrame com as características da nova casa
# As colunas e a ordem devem ser as mesmas usadas para treinar o modelo!
# Use os valores 0 ou 1 para 'existe_segundo_andar' e 'qualidade_da_cozinha_Excelente'
dados_nova_casa = pd.DataFrame([[1, 98, 0, 1]],
                               columns=['quantidade_banheiros', 'area_primeiro_andar',
                                        'existe_segundo_andar', 'qualidade_da_cozinha_Excelente'])

# Assegure-se de que as colunas da nova casa correspondem EXATAMENTE às features usadas no treino
# E que estão na mesma ordem!
try:
    dados_nova_casa_para_prever = dados_nova_casa[available_features]

    # --- Passo 3: Fazer a previsão ---
    preco_estimado = modelo_regressao.predict(dados_nova_casa_para_prever)

    print(f"\nCaracterísticas da nova casa para previsão: {dados_nova_casa_para_prever.iloc[0].to_dict()}")
    print(f"O preço estimado para esta casa é: R$ {preco_estimado[0]:,.2f}")

except KeyError as e:
    print(f"ERRO: A feature '{e}' necessária para a previsão da nova casa não foi usada no treinamento do modelo.")
    print("Verifique se as colunas da 'dados_nova_casa' e 'features_para_modelo' correspondem.")
except Exception as e:
    print(f"Ocorreu um erro inesperado ao preparar a nova casa ou fazer a previsão: {e}")
```

```python
import pandas as pd
import statsmodels.api as sm # Usaremos statsmodels para calcular VIF e modelo
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Caminho do dataset
file_path = "/content/usina.csv"

# Carregar o dataset
try:
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Arquivo não encontrado no caminho especificado: {file_path}")

    df_usina = pd.read_csv(file_path)
    print("Dataset 'usina.csv' carregado com sucesso para análise!")
    print("\nPrimeiras 5 linhas:")
    print(df_usina.head())
    print("\nColunas do dataset:")
    print(df_usina.columns.tolist())

except FileNotFoundError as e:
    print(f"ERRO DE CARREGAMENTO: {e}")
    print("Por favor, verifique se o caminho do arquivo está correto e se ele foi carregado no ambiente do Colab.")
    exit()
except Exception as e:
    print(f"Ocorreu um erro inesperado ao carregar o arquivo: {e}")
    exit()

# Definir a variável resposta (y) e variáveis explicativas (X)
# Assumimos que 'PE' é a variável resposta
y = df_usina['PE']
X = df_usina.drop(columns=['PE']) # Todas as outras colunas são explicativas

# Adicionar uma constante ao X para o cálculo correto do VIF e do modelo Statsmodels
X_const = sm.add_constant(X)

print("\n--- Verificando Multicolinearidade (VIF) ---")
# Calcular o VIF para cada variável explicativa
vif_data = pd.DataFrame()
vif_data["feature"] = X_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_const.values, i)
                   for i in range(len(X_const.columns))]

print(vif_data.round(2))

print("\n--- Construindo o Modelo de Regressão Linear Inicial (Statsmodels) ---")
# Construir o modelo de regressão linear usando Statsmodels para obter um sumário detalhado
modelo_inicial = sm.OLS(y, X_const).fit()
print(modelo_inicial.summary())

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm # Para o modelo_inicial se você rodar essa célula isolada
import pandas as pd # Para o df_usina se você rodar essa célula isolada
import os # Para a verificação do arquivo


# --- Certifique-se de que o df_usina e o modelo_inicial estão definidos ---
# Se você estiver rodando esta célula isoladamente, precisará carregar e treinar o modelo novamente.
# Para evitar erros, incluirei as etapas de carregamento e treino básicas aqui também.

file_path = "/content/usina.csv"

try:
    if 'df_usina' not in locals() or df_usina.empty: # Verifica se df_usina já existe e não está vazio
        if not os.path.exists(file_path):
            raise FileNotFoundError(f"Arquivo não encontrado: {file_path}")
        df_usina = pd.read_csv(file_path)

    y = df_usina['PE']
    X = df_usina.drop(columns=['PE'])
    X_const = sm.add_constant(X)

    if 'modelo_inicial' not in locals(): # Verifica se o modelo já está treinado
        modelo_inicial = sm.OLS(y, X_const).fit()
        print("Modelo de regressão foi (re)treinado para gerar o gráfico de resíduos.")

except FileNotFoundError as e:
    print(f"ERRO: {e}")
    print("Certifique-se de que o arquivo 'usina.csv' está no local correto.")
    exit()
except Exception as e:
    print(f"Ocorreu um erro inesperado ao preparar o ambiente: {e}")
    exit()

# --- Gerar o Gráfico de Resíduos ---
print("\nGerando o Gráfico de Resíduos vs. Valores Previstos...")

# Obter os resíduos do modelo
residuos = modelo_inicial.resid

# Obter os valores previstos pelo modelo
valores_previstos = modelo_inicial.fittedvalues

plt.figure(figsize=(10, 6))
sns.scatterplot(x=valores_previstos, y=residuos, alpha=0.6)
plt.axhline(y=0, color='r', linestyle='--', linewidth=0.8) # Linha de referência no zero
plt.title('Gráfico de Resíduos vs. Valores Previstos')
plt.xlabel('Valores Previstos (PE)')
plt.ylabel('Resíduos')
plt.grid(True, linestyle=':', alpha=0.7)
plt.show()

print("\nGráfico de resíduos gerado! Procure por padrões em forma de 'funil' ou 'cone'.")
```
